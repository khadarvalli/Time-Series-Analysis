When we have data is in stationary format that time we have to use to use `additive`.Because It is good to use when we have linear trend is avalible.

When we have seasonality present inside the data that time we can say there is exponenetial growth is avalible that time we will use the `Multiplicative` model.

```
import pandas as pd
data = pd.read_csv('/content/amazon_revenue_profit.csv',parse_dates=[0])
```

```
data.set_index('Quarter',inplace=True)
ser = data['Revenue']
```

```
ser.plot()
```

```
from statsmodels.tsa.seasonal import seasonal_decompose
```

```
decompose = seasonal_decompose(data['Revenue'],model='multiplicative')
fig = decompose.plot()
```

```
import matplotlib.pyplot as plt
original = decompose.observed
original.plot(figsize=(10,5),title='Original Data Distribution.',color='teal')
plt.show()
trend = decompose.trend
trend.plot(figsize=(10,5),title='Trend Distribution',color='g')
plt.show()
seasonal = decompose.seasonal
seasonal.plot(figsize=(10,5),title='Seasonal Distribution.',color='b')
plt.show()
Residual = decompose.resid
Residual.plot(figsize=(10,5),title='Residual Distribution',color='r')
plt.show()
```

According to the mutiplicative formulae we will varify by using the obsrved values and multiplications of trends,seasonality,Irregularities.

```
original = original[2]
original
```

```
trend[2]*seasonal[2]*Residual[2] #We have got almost same value.
```

For the mutiplicative we are dividing the observed with trend from itself ...for the Additive model we subtract the same.

```
pd.DataFrame(decompose.observed/decompose.trend).plot()
```

In the above way we can detrend our time series data.



**Some Funtions we use in time Series analysis those are:-**

- 1.`diff` - It is take diffrence from future-last with step by default step =1
- 2.`cumsum` - cumulative sum.
- 3.`Groupby` with cetral tendencies.
- 4.`data.index.day`,`data.index.year`,`data.index.month`
- 5 `data.gropuby(data.index.year).mean()`
- 6.`shift`
- 7.`data[:10].pct_change()`  percentage change in the previous to future values.

- 8.`data.expanding.sum()`
- 9.`data.expanding.mean()`  similar to cumulative sum but It is work with diffrent cetral tendecies as well whereas the cumsum only work with sum.

- 10.`data.resample('D')`  daily basis resampleing
- 11.`data.asfreq('2D')` On the frequency of 2 days.
- 12.`data.rolling(window=20).mean()` 20 days rolling.
- 13.`Data.rolling(widow=365).mean()` yearly rolled data.
